```python
class Queue:
    def __init__( self ):
        self.data = [] #Lista vacia

    def is_empty( self ):
        return len( self.data ) == 0

    def length( self ):
        return len( self.data )

    def enqueue( self , value ):
        self.data.append( value ) #Agregas valor a la Q

    def dequeue( self ):        #Quita un valor de la Q
        return self.data.pop(0) # "0" es el primer valor de la linea
    def to_string( self ):
        for elem in self.data:
            print(f"| { elem }| <-- " , end="" )
        print("")
```

```python
cola = Queue()
print("Esta vacio??: " , cola.is_empty())
print("Numero de elementos = " , cola.length() )
cola.enqueue(10)
cola.enqueue(20)
cola.enqueue(30)
print("Esta vacio??: " , cola.is_empty())
print("Numero de elementos = " , cola.length() )
cola.enqueue(40)
cola.enqueue(50)
cola.enqueue(60)
cola.to_string()
siguiente = cola.dequeue()
print("Atendiendo a:" , siguiente)
cola.to_string()
```

#Variantes del ADT Queue
1. Circulares
2. Prioridad
    * Acotada
    * No acotada

```python
#Implementacion de la cola de prioridad acotada

class BoundedPriorityQueue:
    def __init__( self , niveles):
        self.__data = [ Queue() for x in range(niveles) ]
        #slef.__data = Array(niveles)
        #self.__data.clear( Queue() )
        self.__size = 0

    def is_empty( self ):
        return self.__size == 0

    def length( self ):
        return self.__size

    def enqueue( self , prioridad , elem ):
        if prioridad < len( self.__data ) \
            and prioridad >= 0:
            self.__data[ prioridad ].enqueue( elem )
            self.__size +=1

    def dequeue( self ):
        nivel = 0
        while self.__data[nivel].is_empty() \
                and nivel < self.length():
            nivel +=1

        if nivel >= self.length():
            print("Cola esta vacia")
        else:
            return self.__data[ nivel ].dequeue()

    def to_string( self ):
        prioridad = 0
        for cola in self.__data:
            print(f"Prioridad {prioridad} ->", end="")
            if not cola.is_empty():
                cola.to_string()
                prioridad +=1
```

```python
cola = BoundedPriorityQueue(7)

cola.enqueue(1,"Personas con dificultad para moverse.")
cola.enqueue(2,"Personas de la 3ra Edad.")
cola.enqueue(3,"Mujeres embarazadas.")
cola.enqueue(4,"Personas con problemas de salud.")
cola.enqueue(5,"Adultos Normales")
cola.enqueue(6,"Jovenes Adultos")


print("Número de elementos: ",cola.length())
print("Esta vacío?: ",cola.is_empty())
cola.to_string()
print(cola.dequeue())
cola.to_string()
print("Número de elementos=: ",cola.length())
```

```python
#Implementacion de las colas de prioridad sin acotamientos

class PriorityQueue:
    def __init__( self ):
        self.__data =[]
        self.__size = 0

    def is_empty( self ):
        return self.__size == 0

    def length( self ):
        return self.__size

    def enqueue(self , prioridad , elem ):
        existe = False
        indice = 0
        for index in range(len(self.__data)):
            if self.__data[index][ "prioridad" ] == prioridad:
                existe = True
                indice = index
        
        if existe:
            self.__data[indice]["cola"].enqueue(elem)
        else:
            self.__data.append( {"prioridad":prioridad , "cola": Queue() } )
            self.__data[-1]["cola"].enqueue(elem)
    
    def to_string( self ):
        for cola in self.__data:
            cola["cola"].to_string()
    
```

```python
cp = PriorityQueue()
cp.enqueue(1, "Niñas")
cp.enqueue(2, "Niños")
cp.to_string()
```
